# Trigonometry Functions with Numpy and Visualization

Some examples of trigonometry with matplotlib and numpy for learning.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
#import seaborn; seaborn.set()
```

The numpy function linspace is pretty handy for plotting.  Here is divides the range 0 to 2pi in 128 equal segments.

```python
theta = np.linspace(0,2*np.pi,128)
```

```python
s = np.sin(theta)
c = np.cos(theta)
t = np.tan(theta)
```

Stacking up the scatter plots here let's us put it all on one figure.

```python
o = plt.plot(theta,s,c='red')
o = plt.plot(theta,c)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
s1 = s
s2 = np.sin(2*theta)
```

```python
o = plt.plot(theta,s,c='red')
o = plt.plot(theta,s2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```
