# Basic Statistical Distributions with SciPy

Exploring some of the options built into SciPy.  Let's start with importing the necessary modules.

```python
%matplotlib inline
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

SciPy uses objects to represent the distributions.  Let's initialize two normal distributions, both with mean 0.  The first will have a standard deviation of 1, the second will have a standard deviation of 3.

```python
norm1 = st.norm(loc=0.0, scale=1.0)
norm3 = st.norm(loc=0.0, scale=3.0)
```

We need a range of values to plot over.

```python
x = np.linspace(-5,5,100)
```

```python
plt.subplot(211)
plt.plot(x,norm1.pdf(x))
plt.subplot(212)
a= plt.plot(x,norm3.pdf(x))
```

```python
plt.plot(x,norm1.pdf(x))
a= plt.plot(x,norm3.pdf(x))
```

```python
plt.subplot(211)
plt.plot(x,norm1.cdf(x))
plt.subplot(212)
a= plt.plot(x,norm3.cdf(x))
```

```python
plt.plot(x,norm1.cdf(x))
a= plt.plot(x,norm3.cdf(x))
```
