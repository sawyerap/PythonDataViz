# Basic Statistical Distributions with SciPy

Exploring some of the options built into SciPy.  Let's start with importing the necessary modules.

```python
%matplotlib inline
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

SciPy uses objects to represent the distributions.  Let's initialize two normal distributions, both with mean 0.  The first will have a standard deviation of 1, the second will have a standard deviation of 3.

```python
norm1 = st.norm(loc=0.0, scale=1.0)
norm3 = st.norm(loc=0.0, scale=3.0)
```

We need a range of values to plot over.

```python
x = np.linspace(-5,5,100)
```

## Normal Distributions

```python
plt.subplot(211)
plt.plot(x,norm1.pdf(x))
plt.subplot(212)
a= plt.plot(x,norm3.pdf(x))
```

```python
plt.plot(x,norm1.pdf(x))
a= plt.plot(x,norm3.pdf(x))
```

```python
plt.subplot(211)
plt.plot(x,norm1.cdf(x))
plt.subplot(212)
a= plt.plot(x,norm3.cdf(x))
```

```python
plt.plot(x,norm1.cdf(x))
a= plt.plot(x,norm3.cdf(x))
```

## Weibull Distributions

The Weibull extreme value distribution has a shape parameter, called <i>c</i> in the documentation.  It also has a scale term (which flattens the curve) and a shift (loc) term.

```python
w1 = st.weibull_min(1.79)
w2 = st.weibull_min(1.79, scale = 2.0)
w3 = st.weibull_min(1.79, scale = 1.0, loc = 1.0)
w4 = st.weibull_min(1.79, scale = 2.0, loc = 1.0)
x = np.linspace(0,10,100)
```

```python
plt.plot(x,w1.pdf(x))
plt.plot(x,w2.pdf(x))
plt.plot(x,w3.pdf(x))
plt.plot(x,w4.pdf(x))
a= plt.legend(['Base','Scale = 2','Scale = 1, Loc = 1','Scale = 2, Loc = 1'])
```

## Gamma Distribution

Let's use the gamma to learn some more functionality in scipy.stats.
gamma.pdf(x, a, loc, scale)

```python
g1 = st.gamma(1.75)
g2 = st.gamma(1.75, loc=2.0)
g3 = st.gamma(1.75, scale=2.0)
```

Let's use the ppf method (Percent point function) to determine good bounds for our linear space.  Percent point function is the inverse of the CDF.

```python
g1.ppf(0.01)
```

```python
g1.ppf(0.999)
```

```python
x = np.linspace(0,10,100)
```

```python
plt.plot(x,g1.pdf(x))
plt.plot(x,g2.pdf(x))
plt.plot(x,g3.pdf(x))
a= plt.legend(['Base','Loc = 2','Scale = 2'])
```
