# Problems with pandas and date index

I had a lot of problems getting a single row out of a date indexed dataframe.  I was trying to access the row by `dataframe['2020-02-14']` but kept getting a keyerror.  Slices would work, `dataframe['2020-02-01':'2020-02-14']`.  I needed to use the .loc method as shown below.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
```

```python
date_range = pd.date_range(dt.date(2020,1,1),dt.date(2020,4,1), freq='D')
```

```python
date_range
```

```python
ds1 = np.random.randn(len(date_range))
ds2 = 10 * np.random.randn(len(date_range))
```

```python
edf = pd.DataFrame(None, index=date_range)
edf['ds1'] = ds1
edf['ds2'] = ds2
```

```python
edf
```

```python
edf['2020-02']
```

```python
edf['2020-02-15']
```

```python
edf['2020-02-15':'2020-02-20']
```

```python
edf['2020-02-15':'2020-02-15']
```

```python
edf.loc['2020-02-15']
```
