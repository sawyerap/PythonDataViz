# Merge/Join Operations in Pandas

Examples and trials with merge and join.  First we need to import the usual and get some data.

```python
import numpy as np
import pandas as pd
```

```python
a = ['apple', 'orange', 'broccoli', 'banana', 'starfruit', 'cherry', 'strawberry', 'tomato', 'egg']
c = ['red', 'orange', 'green', 'yellow', 'yellow', 'red', 'red', 'red', 'white']
d = np.random.randint(0,10,9)
g = [1, 1, 2, 1, 1, 1, 1, 2, 3]
```

```python
df1 = pd.DataFrame({'gitem': a, 'color': c, 'data1': d, 'gtype': g})
```

```python
df2 = pd.DataFrame({'gtype': [1, 2, 3], 
              'category': ['by_count', 'by_weight', 'by_count'],
              'tracking': [45, 78, 99]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
# Best practice is to specify
pd.merge(df1, df2, on='gtype')
# If key names don't match, don't worry, left_on='lkey', right_on='rkey')
# also a 'how' with  'inner' (default, 'left', 'right', and 'outer'
```
