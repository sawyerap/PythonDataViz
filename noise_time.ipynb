# Noise Time Series



```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
init_x = 4
```

```python
values = 2*[init_x]
```

```python
def dnorm(mean, variance, size=1):
    if isinstance(size, int):
        size = size,
    return mean + np.sqrt(variance) * np.random.randn(*size)
```

```python
N = 1000
```

```python
b0 = 0.8
b1 = -0.4
```

```python
noise = dnorm(0, 0.1, N)
```

```python
for i in range(N):
    new_x = values[-1] * b0 + values[-2] * b1 + noise[i]
    values.append(new_x)
```

```python
plt.plot(values)
```
