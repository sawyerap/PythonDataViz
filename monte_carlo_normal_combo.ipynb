```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

## Monte Carlo Examination of Using Normal Curves

Tests if the average of 10 normal distributions (standard deviation 1.0) matches expectations of standard devation 1/sqrt(10) 

```python
norm1 = st.norm(loc=0.0, scale=1.0)
norm2 = st.norm(loc=0.0, scale=1/np.sqrt(10))
```

```python
nsim = 50000
ndraws = 10
results = np.zeros(nsim)
for i in range(nsim):
    data = norm1.rvs(ndraws)
    avg = np.average(data)
    results[i] = avg
```

```python
edge, dens = np.histogram(results, bins=50, density=True)
```

```python
x = np.linspace(-5,5,100)
#
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8))
ax.plot(x,norm1.pdf(x), label='g1')
ax.plot(x,norm2.pdf(x), label='g2')
ax.hist(results, bins=50, density=True)
ax.grid()
```
